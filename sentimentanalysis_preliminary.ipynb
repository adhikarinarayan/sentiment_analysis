## VADER

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sa = SentimentIntensityAnalyzer()
```

```python
import pandas as pd
```

```python
df= pd.read_csv('cleanedv3.csv')
```

```python
df = df.sample(frac=.90)
```

```python
df
```

```python
df1 = df.prompt[:20]
```

```python
df1 = pd.DataFrame(df1)
```

```python
def compound_pol(f):
  a = sa.polarity_scores(f)
  return a['compound']
```

```python
sa.polarity_scores
```

```python
df1
```

```python
df1['vader']=df1['prompt'].apply(compound_pol)
```

```python
df1.head(20)
```

## TextBlob

```python
from textblob import TextBlob
```

```python
df2 = df1
```

```python
df2
```

```python
df2 = pd.DataFrame(df2)
```

```python
def getPolarity(text):
   return round(TextBlob(text).sentiment.polarity,2)
```

```python
df2['Textblob']=df2['prompt'].apply(getPolarity)#('I feel guilty for not being able to protect')
```
